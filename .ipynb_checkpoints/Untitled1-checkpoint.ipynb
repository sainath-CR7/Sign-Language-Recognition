# Importing the Keras libraries and packages

```python
import tensorflow as tf
import cv2
from tensorflow import keras
from tensorflow.keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import SGD
```

# Importing and preprocessing the dataset

```python
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1. / 255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1. / 255)

training_set = train_datagen.flow_from_directory(r'C:\Users\sai nath rathod\anaconda3\project-MB\mydata\training_set',target_size=(64, 64),batch_size=32,class_mode='categorical')

test_set = test_datagen.flow_from_directory(r'C:\Users\sai nath rathod\anaconda3\project-MB\mydata\test_set',target_size=(64, 64),batch_size=32,class_mode='categorical')
```

# BUILDING THE CNN

```python
# Initialing the CNN
classifier = Sequential()

# Step 1 - Convolution Layer
classifier.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(64, 64, 3)))

# step 2 - Pooling
classifier.add(MaxPool2D(pool_size=(2, 2), strides=2))

# Adding second convolution layer
classifier.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu', padding='same'))
classifier.add(MaxPool2D(pool_size=(2, 2), strides=2))

# Adding 3rd Concolution Layer
classifier.add(Conv2D(filters=128, kernel_size=(3, 3), activation='relu', padding='valid'))
classifier.add(MaxPool2D(pool_size=(2, 2), strides=2))

# Step 3 - Flattening
classifier.add(Flatten())

# Step 4 - Full Connection
classifier.add(Dense(256, activation='relu'))
classifier.add(Dropout(0.5))
classifier.add(Dense(26, activation='softmax'))

# Compiling The CNN
classifier.compile(
    optimizer=SGD(learning_rate=0.01),
    loss='categorical_crossentropy',
    metrics=['accuracy'])
```

```python
classifier.summary()
```

# Fitting the Model

```python
model = classifier.fit_generator(
    training_set,
    steps_per_epoch=800,
    epochs=35,
    validation_data=test_set,
    validation_steps=6500
)
```

```python
model.history
```

```python
# saving the model
classifier.save('monu.keras')
```
